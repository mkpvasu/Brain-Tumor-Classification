```python
df = df.drop(['transect_id','beach_geomorphology'], axis = 1)
```

```python
d = {'m_surf': 'sum', 'm_wrack': 'sum', 'm_bare': 'sum', 'm_wood': 'sum', 'm_vegetation': 'sum', 'm_total': 'sum', 'surf_count': 'sum',
     'wrack_count': 'sum', 'bare_count': 'sum', 'wood_count': 'sum', 'vegetation_count': 'sum', 'total_debris_count': 'sum'}
df_new = df.groupby(['survey_id','beach_id','survey_date','year','month','day','beach_name','beach_latitude','beach_longitude','beach_width',
                              'beach_length','beach_substrate','beach_orientation'], as_index=False).agg(d).reindex(columns=df.columns)
```

```python
df_new = df_new.groupby(['beach_name']).filter(lambda s: s.year.isin([2019, 2020, 2021]).sum() >= 6)
#df_new
```

```python
df_new = df_new.merge(df_full.groupby('survey_id')['is_floppy', 'has_loops'].sum().reset_index(), on='survey_id', how='left')
```

```python
df_filtered = df_new.drop(['survey_id','survey_date','month','day','m_surf','m_wrack','m_bare','m_wood','m_vegetation','m_total'],axis=1)
#df_filtered.info()
```

```python
d = {'surf_count': 'sum', 'wrack_count': 'sum', 'bare_count': 'sum', 'wood_count': 'sum', 
     'vegetation_count': 'sum', 'total_debris_count': 'sum',
     'is_floppy': 'sum', 'has_loops': 'sum'}
df_filtered = df_filtered.groupby(['beach_id','year','beach_name','beach_latitude','beach_longitude','beach_width',
                                            'beach_length','beach_substrate','beach_orientation'], as_index=False).agg(d).reindex(columns=df_filtered.columns)
#df_filtered
```

```python
df_filtered = df_filtered.assign(avg_debris_density = lambda s:s.total_debris_count/(s.beach_length*1000))
df_filtered = df_filtered.round({'avg_debris_density': 3})

df_filtered = df_filtered.assign(avg_entanglement_debris_density = lambda s:((s.is_floppy)+(s.has_loops))/(s.beach_length*1000))
df_filtered = df_filtered.round({'avg_entanglement_debris_density': 3})

df_filtered['entanglement fraction'] = (df_filtered['is_floppy']/df_filtered['total_debris_count']).round(3)
```
